```
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd

cancer = load_breast_cancer()

data = cancer.data[:]
target = cancer.target

table = []

for i in range (0, len(data)):
  t = {
      'Radius' : data[i][0],
      'Texture' : data[i][1],
      'Perimeter' : data[i][2],
      'Area' : data[i][3],
      'Smoothness' : data[i][4],
      'Compactness' : data[i][5],
      'Concavity' : data[i][6],
      'Concave Point' : data[i][7],
      'Symmetry' : data[i][8],
      'Fracral Dimension' : data[i][9],
      'Diagnosis' : target[i],
      }
  table.append(t)
df = pd.DataFrame(table)
df
```

Note : <br> Diagnosis <br>
0 = Malignant <br>
1 = Benign



```
#Contoh penggunaan mean radius, mean perimeter, mean compactness
data = data[:,[0, 2, 5]]
```

```
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
#gini impurity
x = data 
y = target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)

clf = DecisionTreeClassifier(max_depth=5) #Manipulasi akurasi di sini
clf = clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
print(y_pred)
```

```
from sklearn import metrics

metric = metrics.accuracy_score(y_test, y_pred)
print("\nAccuracy : ", metric)
```

```
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus

feature_cols = ['Radius', 'Perimeter', 'Compactness',]

dot_data = StringIO()
export_graphviz(
    clf, out_file=dot_data,
    filled=True, rounded = True,
    special_characters = True,
    feature_names = feature_cols,
    class_names = ['0', '1']
    )

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('Breast_Cancer.png')
Image(graph.create_png())
```

```
#entropy impurity
clf = DecisionTreeClassifier(criterion = 'entropy', max_depth=5) #Manipulasi akurasi di sini
clf = clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
print(y_pred)
```

```
metric = metrics.accuracy_score(y_test, y_pred)
print("\nAccuracy : ", metric)
```

```
feature_cols = ['Radius', 'Perimeter', 'Compactness',]

dot_data = StringIO()
export_graphviz(
    clf, out_file=dot_data,
    filled=True, rounded = True,
    special_characters = True,
    feature_names = feature_cols,
    class_names = ['0', '1']
    )

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('Breast_Cancer.png')
Image(graph.create_png())
```

```
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer

df = pd.DataFrame(x)
df
```

```
x = df.values

imputer = SimpleImputer()
imputedData = imputer.fit_transform(x)

scaler = MinMaxScaler(feature_range=(0,1))
normalizedData = scaler.fit_transform(imputedData)
```

```
from sklearn.ensemble import BaggingClassifier
from sklearn import model_selection

x = normalizedData[:]
y = target

kfold = model_selection.KFold(n_splits = 100, random_state = 100, shuffle=True)
clf = DecisionTreeClassifier()

num_trees = 100

model = BaggingClassifier(base_estimator=clf, n_estimators=num_trees, random_state=4)
results = model_selection.cross_val_score(model, x, y, cv=kfold)
print(results.mean())
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier

kfold = model_selection.KFold(n_splits=100, random_state=100, shuffle=True)

estimation = []
model_1 = LogisticRegression()
estimation.append(('LogisticRegretion', model_1))

model_2 = DecisionTreeClassifier()
estimation.append(('DecisionTree', model_2))

model_3 = SVC()
estimation.append(('SVM', model_3))

ensamble = VotingClassifier(estimation)
result = model_selection.cross_val_score(ensamble, x, y, cv=kfold)
print(result.mean())
```
